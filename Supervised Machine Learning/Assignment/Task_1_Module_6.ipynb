## Task-1 : Linear Regression for House Price Prediction.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("BostonHousing.csv")
```

```python
data
```

```python
data.rename(columns={"medv":"price"},inplace=True)
```

```python
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['rm'].fillna(data['rm'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Distribution of MEDV')
plt.show()
```

```python
corr = data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x='rm', y='price', data=data)
plt.title('Rooms vs House Price')
plt.show()
```

```python
sns.scatterplot(x='lstat', y='price', data=data)
plt.title('LSTAT vs House Price')
plt.show()
```

```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
print("X-Shape :",x.shape)
print("Y-Shape :",y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
mse = mean_squared_error(y_pred,y_test)
print("Mean Squared Error:",mse)
```

```python
r2 = r2_score(y_pred,y_test)
print("R2 Score :",r2)
```
